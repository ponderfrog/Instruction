```python
# 类的标准写法

from typing import Any


class Student(object):
    # 限制实例可拥有的属性
    # 元组内包含允许的属性
    # 仅对当前类起作用（子类不起作用）
    # 子类中也定义slots后，子类允许的属性包括子类中的元组内容加上父类
    __slots__ = ('name','__score')

    # print()输出的实例信息
    def __str__(self):
        return 'this is str'

    # 实例返回的信息
    # 可以__repr__ = __str__
    def __repr__(self):
        return f'hello world'
    
    # 结合__next__()方法使用
    def __iter__(self):
        return self
    def __next__(self):
        # 条件仅做示意
        if self.name:
            raise StopIteration()
        return self.name
    
    # 实现实例[n]取元素功能或切片功能
    def __getitem__(self, n):
        # 仅做示意
        return n
    
    # __setitem__() 视作list或dic进行赋值操作
    # __delitem__() 删除某个元素

    # 没有属性时进入该方法查找
    # 仍没有时输出None
    def __getattr__(self, attr):
        if attr == 'age':
            return 24
        # 手动抛出错误，否则默认为None
        raise AttributeError(f'xxx')

    # 实例直接加()调用时进入该函数
    def __call__(self):
        pass

    # 并不是构造函数
    # 初始化方法
    def __init__(self, name, score) -> None:
        self.name = name
        # 私有变量
        # 仅限内部访问
        # _Student__score仍然可以访问（强烈不推荐）
        self.__score = score
    def print_score(self):
        print(f'{self.name}\'s score is {self.__score}')

# 实例
bart = Student('bart Simpson', 90)
bart
```

```python
# 继承与多态

class Animal(object):
    def run(self):
        print('Animal is running...')

# Dog类同时是Animal类
# Animal类不是Dog类
class Dog(Animal):
    # 覆盖父类方法
    def run(self):
        print('Dog is running...')
    
    def eat(self):
        print('Dog is eating meat...')

# 多态
# 传入Animal类则调用父类方法
# 传入Dog类则调用子类方法
# 只要类型有run()方法就可以传入（动态语言鸭子类型）
def run_2(animal):
    animal.run()
```

```python
# @property方法
# 直接通过方法名来访问方法，不需要()
# 一般作用于不对外暴露的变量

class Student(object):

    # 注意方法名和变量名不能相同
    # 加上property即实现了getter方法
    @property
    def birth(self):
        return self._birth

    # 如果没有定义setter则成为只读变量
    # 方法名+setter
    @birth.setter
    def birth(self, value):
        self._birth = value
    
    # 清除属性
    @birth.deleter
    def birth(self):
        self._area = 0
```

```python
# 枚举类
from enum import Enum, unique

# 自动给member从1开始赋值
Month = Enum('Month', ('Jan', 'Feb', 'Mar', 'Apr', 'May'))
for name, member in Month.__members__.items():
    print(name, member,member.value)


#@unique装饰器检查没有重复值
@unique
class Weekday(Enum):
    # 指定member值
    Sun = 0
    Mon = 1
    Tue = 2
    Wed = 3
    Thu = 4
    Fri = 5
    Sat = 6
# 访问的方法
day1 = Weekday.Mon

Weekday['Mon']
Weekday(1)

for name, member in Weekday.__members__.items():
    print(name, member, member.value)
```

少见，不常使用

```python
# __new__详解

# __new__除cls外的参数如何理解？？？

class demo(object):
    # 真正的类构造函数
    # 第一个参数是当前正在实例化的类，其余参数会传递给__init__方法进行初始化
    def __new__(cls):
        print('__new__')
        # 返回一个实例化对象
        # 如果返回None则不会继续运行__init__
        # 返回值即__init__中的self
        return super().__new__(cls)
        # return None

d = demo()
```

```python
# Metaclass

# 元类的命名以Metaclass结尾，以表明其身份
# Metaclass从type类派生
class AddInfoMetaclass(type):
    def __new__(cls, name, bases, attr):
        attr['info'] = 'add by metaclass'
        return super().__new__(cls, name, bases, attr)

class demo(metaclass = AddInfoMetaclass):
    pass
```

```python
# 错误处理
# 所有错误继承自BaseException

try:
    a = 10/0
# 可以有多个except
# 捕获不同类型的错误
except ValueError as e:
    print('error!')
except ZeroDivisionError as e:
    raise
# 无错误时执行else
else:
    print('no error')
# 无论是否出错最后执行finally
# 即finally一定执行
finally:
    print('End')


# 自定义错误
# 尽量使用python内置的错误类型
class MyError(ValueError):
    pass
# raise后面可以不接内容
raise MyError(f'invalid value')

# 用assert代替print查看错误信息
# 若判断式结果不为True，则抛出逗号后的错误信息（AssertionError）
# python解释器中用-O（英文大写字母O）关闭assert
# 关闭后assert等于pass
assert 10, 'error'

# logging
import logging
# 确定debug, info, warning, error等错误等级
logging.basicConfig(level=logging.INFO)
logging.info()
```
